```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data=data[["iyear","imonth","iday","country_txt","region_txt","provstate","city",
       "latitude","longitude","location","summary","attacktype1_txt","targtype1_txt",
       "gname","motive","weaptype1_txt","nkill","nwound","addnotes"]]
data.head()
```

```python
data.rename(columns =
    {"iyear":"Year","imonth":"Month","iday":"Day","country_txt":"Country",
                   "region_txt":"Region","provstate":"Province/State","city":"City",
                   "latitude":"Latitude","longitude":"Longitude","location":"Location",
                   "summary":"Summary","attacktype1_txt":"Attack Type","targtype1_txt":"Target Type",
                   "gname":"Group Name","motive":"Motive","weaptype1_txt":"Weapon Type",
                   "nkill":"Killed","nwound":"Wounded","addnotes":"Add Notes"
},inplace=True)

data.head()
```

```python
data.isnull().sum()
```

```python
data["Killed"]=data["Killed"].fillna(0)
data["Wounded"]=data["Wounded"].fillna(0)
data["Casualty"]=data["Killed"]+data["Wounded"]
```

# YEARWISE ATTACK

ATTACKS IN EACH YEAR

```python
attacks=data["Year"].value_counts(dropna=False).sort_index().to_frame().reset_index().rename(columns={"index":"Year","Year":"Attacks"}).set_index("Year")
attacks.head()
```

```python
attacks.plot(kind="bar",color="red" )
plt.title("Timeline of Attacks",fontsize=15)
plt.show()
```

 Most number of attacks(16903) in 2014


 Least number of attacks(471) in 1971

casualities in each year

```python
cas=data[["Year","Casualty"]].groupby("Year").sum()
cas.head()
```

```python
cas.plot(kind="bar",color="red" )
plt.show()
```

```python
kil=data[["Year","Killed"]].groupby("Year").sum()
kil.head()
```

```python
wo=data[["Year","Wounded"]].groupby("Year").sum()
wo.head()
```

```python
kil.plot(kind="bar",color="red" )
plt.show()
wo.plot(kind="bar",color="orange" )
plt.show()
```

# REGION WISE ATTACK

Distribution of Terrorist Attacks over Regions from 1970-2017

```python
reg= pd.crosstab(data.Year,data.Region)
reg.head()
```

```python
reg.plot(kind="area" , stacked= False)
plt.show()
```

Total Terrorist Attacks in each Region from 1970-2017

```python
regt=reg.transpose()
regt["Total"]=regt.sum(axis=1)
ra=regt["Total"].sort_values(ascending=False)
ra
```

```python
ra.plot(kind="bar",color="red" )
plt.show()
```
